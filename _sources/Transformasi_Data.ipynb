## Proses Menyatukan database menjadi satu

```python
import pandas as pd
import psycopg2
import pymysql

# -----------------------
# Ambil data dari PostgreSQL
# -----------------------
conn_pg = psycopg2.connect(
    host='localhost',
    database='postgres',
    user='postgres',
    password='malik123'
)

query_sepal = "SELECT id, class, sepal_length, sepal_width FROM sepal_postgree"
df_sepal = pd.read_sql(query_sepal, conn_pg)
conn_pg.close()

# -----------------------
# Ambil data dari MySQL
# -----------------------
conn_mysql = pymysql.connect(
    host='localhost',
    user='root',
    password='',
    database='petal_mysql'
)

query_petal = "SELECT id, petal_length, petal_width FROM petal"
df_petal = pd.read_sql(query_petal, conn_mysql)
conn_mysql.close()
```

Kode ini digunakan untuk mengambil data dari dua basis data yang berbeda  PostgreSQL dan MySQL  dengan memanfaatkan Python dan library pandas.

Pada bagian pertama, skrip membuka koneksi ke PostgreSQL melalui library psycopg2, lalu menjalankan query untuk mengambil kolom id, class, sepal_length, dan sepal_width dari tabel sepal_postgree. Data hasil query kemudian dimasukkan ke dalam DataFrame pandas bernama df_sepal agar mudah dianalisis maupun diolah lebih lanjut. Setelah proses selesai, koneksi ke PostgreSQL ditutup supaya tidak memboroskan sumber daya.

Selanjutnya, bagian kedua melakukan langkah serupa untuk basis data MySQL menggunakan library pymysql. Koneksi dibuat ke database petal_mysql, kemudian dieksekusi query yang mengambil kolom id, petal_length, dan petal_width dari tabel petal. Data yang diperoleh disimpan dalam DataFrame df_petal untuk memudahkan manipulasi dan analisis di Python. Dengan cara ini, informasi dari dua sistem basis data yang berbeda dapat dikelola, digabungkan, atau dianalisis secara terpisah, sehingga mendukung integrasi data dari berbagai sumber.

```python
import pandas as pd
from sqlalchemy import create_engine

# PostgreSQL
engine_pg = create_engine("postgresql+psycopg2://postgres:malik123@localhost/postgres")
df_sepal = pd.read_sql("SELECT id, class, sepal_length, sepal_width FROM sepal_postgree", engine_pg)

# MySQL
engine_mysql = create_engine("mysql+pymysql://root:@localhost/petal_mysql")
df_petal = pd.read_sql("SELECT id, petal_length, petal_width FROM petalmysql", engine_mysql)

df_merged = pd.merge(df_sepal, df_petal, on='id', how='inner')

for col in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    df_merged[col] = df_merged[col].astype(float)
```

Kode ini digunakan untuk mengambil sekaligus menggabungkan data yang tersimpan di dua basis data berbeda, yaitu PostgreSQL dan MySQL, dengan memanfaatkan Python, library pandas, dan SQLAlchemy.

Langkah pertama adalah membuat engine koneksi ke PostgreSQL menggunakan fungsi create_engine, lalu menjalankan query untuk mengambil kolom id, class, sepal_length, dan sepal_width dari tabel sepal_postgree. Data yang didapat disimpan dalam DataFrame df_sepal. Setelah itu, dibuat engine koneksi ke MySQL dengan cara yang sama, kemudian dieksekusi query untuk mengambil kolom id, petal_length, dan petal_width dari tabel petalmysql dan menyimpannya pada DataFrame df_petal. Dengan pendekatan ini, data dari dua sumber yang berbeda dapat diakses dengan format seragam berupa DataFrame di Python.

Berikutnya, kedua DataFrame tersebut digabungkan melalui operasi merge berdasarkan kolom id menggunakan metode inner join. Proses ini menghasilkan DataFrame df_merged yang hanya berisi baris dengan id yang ada di kedua tabel. Selanjutnya, kolom numerik seperti sepal_length, sepal_width, petal_length, dan petal_width diubah tipenya menjadi float agar siap untuk dianalisis atau dihitung lebih lanjut. Cara ini membuat integrasi data dari PostgreSQL dan MySQL menjadi lebih praktis serta siap digunakan untuk pemrosesan data di Python.
